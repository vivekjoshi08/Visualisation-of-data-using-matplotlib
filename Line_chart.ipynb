#### Custom Line Color

To change the line color, use the **‘color’** argument.

```python
df.loc['Norway'].plot(color = 'orange')
plt.title('GDP of European country', color = 'white', size = 20)
plt.xlabel('Year', color = 'white', size =15)
plt.ylabel('GDP per capital', color = 'white', size = 15)
plt.xticks(color = 'white', rotation = 45)
plt.yticks(color = 'white')
plt.legend(loc = 'best')
```

You can also add transparency to the color with the **'alpha'** argument (0=transparent, 1=opaque)

```python
df.loc['Norway'].plot(color = 'orange', alpha = 0.5)
plt.title('GDP of European country', color = 'white', size = 20)
plt.xlabel('Year', color = 'white', size =15)
plt.ylabel('GDP per capital', color = 'white', size = 15)
plt.xticks(color = 'white', rotation = 45)
plt.yticks(color = 'white')
plt.legend(loc = 'best')
```

#### Custom Line Style

You can choose between different line styles with the **'linestyle'** argument.
A few of the line styles available are:
- **-**
- **--**
- **-.**
- **:**


We will now see how these different line styles look like in a line plot.

```python
df.loc['Norway'].plot(color = 'orange', alpha = 0.8, linestyle='-')
plt.title('GDP of European country', color = 'white', size = 20)
plt.xlabel('Year', color = 'white', size =15)
plt.ylabel('GDP per capital', color = 'white', size = 15)
plt.xticks(color = 'white', rotation = 45)
plt.yticks(color = 'white')
plt.legend(loc = 'best')
```

```python
df.loc['Norway'].plot(color = 'orange', alpha = 0.8, linestyle='--')
plt.title('GDP of European country', color = 'white', size = 20)
plt.xlabel('Year', color = 'white', size =15)
plt.ylabel('GDP per capital', color = 'white', size = 15)
plt.xticks(color = 'white', rotation = 45)
plt.yticks(color = 'white')
plt.legend(loc = 'best')
```

```python
df.loc['Norway'].plot(color = 'orange', alpha = 0.8, linestyle='-.')
plt.title('GDP of European country', color = 'white', size = 20)
plt.xlabel('Year', color = 'white', size =15)
plt.ylabel('GDP per capital', color = 'white', size = 15)
plt.xticks(color = 'white', rotation = 45)
plt.yticks(color = 'white')
plt.legend(loc = 'best')
```

```python
df.loc['Norway'].plot(color = 'orange', alpha = 1, linestyle=':')
plt.title('GDP of European country', color = 'white', size = 20)
plt.xlabel('Year', color = 'white', size =15)
plt.ylabel('GDP per capital', color = 'white', size = 15)
plt.xticks(color = 'white', rotation = 45)
plt.yticks(color = 'white')
plt.legend(loc = 'best')
```

#### Custom Line width

You can custom the line width usng the **'linewidth'** argument.

```python
df.loc['Norway'].plot(color = 'orange', alpha = 0.8, linestyle='--', linewidth= 5)
plt.title('GDP of European country', color = 'white', size = 20)
plt.xlabel('Year', color = 'white', size =15)
plt.ylabel('GDP per capital', color = 'white', size = 15)
plt.xticks(color = 'white', rotation = 45)
plt.yticks(color = 'white')
plt.legend(loc = 'best')
```

#### Add annotation in the line plot

```python
df.loc['Norway'].plot(color = 'orange', alpha = 0.8, linestyle='-.')
plt.title('GDP of European country', color = 'white', size = 20)
plt.xlabel('Year', color = 'white', size =15)
plt.ylabel('GDP per capital', color = 'white', size = 15)
plt.xticks(color = 'white', rotation = 45)
plt.yticks(color = 'white')
plt.legend(loc = 'best')
plt.annotate('GDP of NORWAY', xy= (1965, 40000), size = 20, color = 'cyan')
```

Now we will add the observations on top of each point plotted in the line chart.
Consider the following example where we plot values generated from two populated lists.

```python
df_norway = df.loc['Norway']
df_norway.plot(color ='red', linewidth = 2)
df_norway
```

```python
xs = [1,2,3,4,5,6,7,8]
ys = [10,20,30,40,50,60,70,80]

plt.plot(xs,ys, 'bo-')

for x,y in zip(xs,ys):
    label = (x,y)
    plt.annotate(label,                           # this is the text
                 (x,y),                           # this is the point to label
                 textcoords="offset points",      # how to position the text
                 xytext=(0,10),                   # distance from text to points (x,y)
                 ha='center')                     # horizontal alignment can be left, right or center
```
