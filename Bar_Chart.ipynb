#### panda Basics

We will now import the **pandas** module.

```python
import pandas as pd
```

- Before plotting, we convert the column headings from a string to integer data type, since they represent numerical values.
- Extract year from last 4 characters of each column name.
- The current column names are structured as 'gdpPercap_(year)', 
- so we want to keep the (year) part only for clarity when plotting GDP vs. years
- To do this we use strip(), which removes from the string the characters stated in the argument
- This method works on strings, so we call str before strip()

```python
df_europe = pd.read_csv(r'C:\Users\HP\Desktop\MachineLearning\Visualisation_of_data\Dataset\europe.csv', index_col = 'country')
df_europe.drop('Total', axis = 1, inplace = True)

df_europe.head(5)

```

```python
years = df_europe.columns.str.strip('gdpPercap_')

df_europe.columns = years.astype(int)

print(df_europe.columns)
```

```python
df_europe.plot(kind = 'bar')
plt.show()
```

```python
df_europe.loc['Austria'].plot(kind='bar')
plt.show()
```

```python
df_europe.iloc[0:2].plot(kind='bar')
plt.show()
```

```python
df_europe.iloc[0:2, 0:5].plot(kind='bar')
plt.show()
```

```python
df_europe.loc["Austria"].plot(kind ='bar', edgecolor ='black', color = 'cyan')
plt.title('GDP of Europian countries over time', color = 'white', size = 50)
plt.xlabel('GDP per acpital', color = 'white', size = 20)
plt.ylabel('Amount in Euros (millions)', color = 'white', size = 20)
plt.xticks(color='white', rotation = 90)
plt.yticks(color = 'white')
plt.rc( 'axes' , titlesize = 10)
```

```python
df_europe.loc['Austria'].plot(kind='bar', hatch =  '*', color = 'blue')
plt.show()
```

```python
df_Austria = df_europe.loc['Austria']
print(df_Austria)
```

```python
df_Austria.plot(kind='barh', color = 'blue', hatch ='*')
plt.xticks(rotation = 45, color = 'white')
plt.yticks(color = 'white')
for index,value in enumerate(df_Austria):
    
    label = format(int(value), ',')
    
    plt.annotate( label, xy= (value, index - 0.20), color = 'red')
```

We can also plot the same bar graph vertically using style sheets available. Use ***style.use()***

```python
import matplotlib as mpl
import matplotlib.pyplot as plt

mpl.style.use('ggplot')

ax = df_Austria.plot(kind = 'bar', figsize = (12,12), width = 0.6)

plt.title("Gdp of Austria in past few years")

x_offset = 0
y_offset = 0

for p in ax.patches:
    b = p.get_bbox()
    val = '{:.2f}'.format(b.y1 + b.y0)
    ax.annotate(val, ((b.x0 + b.x1)/2 + x_offset, b.y1 + y_offset))
plt.show()
```
